```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv');
movies=pd.read_csv('tmdb_5000_movies.csv');
```

```python
movies.head(1)
```

```python
credits.shape
```

```python
movies= movies.merge(credits,on='title')
```

```python
movies.head(1)
```

# noting important features
# generes, id, keywords ,title, overview ,cast, crew

```python
movies=movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
import ast
ast.literal_eval
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# with this function we take the names form genres as list 
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert) 
# similarly we convert key words
```

```python
movies.head()
```

```python
 def convert3(obj):
     L= []
     counter=0
     for i in ast.literal_eval(obj):
         if counter !=3:
             L.append(i['name'])
             counter+=1
         else:
             break
     return L
```

```python
movies['cast' ]= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
        break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']= movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
movies.head()
```

```python
# here we make a space words to one
movies['genres']= movies['genres'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['keywords']= movies['genres'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['cast']= movies['genres'].apply(lambda x:[ i.replace(" ","") for i in x])
movies['crew']= movies['genres'].apply(lambda x:[ i.replace(" ","") for i in x])
```

```python
# we join the following 
movies['tags']= movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df =movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
 new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
# stamming is  a technique used here where we have similary words like [love, loving,loved] 
# to just [love,love,love]
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# now we will vectorize the tags ..we will choose the colsest  to its  vectors
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

# now its almost done ..now we will find the distance of vectors form ecah other by cosine distance 
# not the eculedian dist

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#  now we have to sort such that index dont loose while sorted ..so we use enum and list 
# and lambda to sort the second thing in the list not the index
```

```python
similarity= cosine_similarity(vectors)
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

# lets do by knn as well

```python
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# Assuming 'vectors' is your list of vectors
# Compute cosine similarity matrix
similarity = cosine_similarity(vectors)

# Plot the cosine similarity matrix
plt.imshow(similarity, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Cosine Similarity')
plt.title('Cosine Similarity Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the frequency of each genre
genre_counts = {}
for genres_list in movies['genres']:
    for genre in genres_list:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1

# Sort the genres by frequency
sorted_genres = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar([genre[0] for genre in sorted_genres], [genre[1] for genre in sorted_genres])
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Genres')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] 
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
         # print (i[0])  will give the id 
        print(new_df.iloc[i[0]].title) # while this will give the title related to id
```

```python
recommend('Furious 7')
```
